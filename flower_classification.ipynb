# 1-import necessary libraries 

```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from tensorflow.keras import layers,models
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

# 2-Load data

```python
data=[]
labels=[]

for item in glob.glob("/content/17flowers/*/*"):
    img=cv2.imread(item)
    re_img=cv2.resize(img,(32,32))
    label=item.split("/")[-2]
    data.append(re_img)
    labels.append(label)
data=np.array(data)/255
le=LabelEncoder()
labels=le.fit_transform(labels)
Labels=to_categorical(labels,17)
xtrain,xtest,ytrain,ytest=train_test_split(data,labels,test_size=0.2)
```

# 3-Build my own network

```python
network=models.Sequential([
    layers.Conv2D(32,(3,3),activation="relu",input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.Conv2D(32,(3,3),activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64,(3,3),activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(32,(3,3),activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(100,activation="relu"),
    layers.Dropout(0.4),
    layers.Dense(17,activation="softmax")
])
```

```python
network.compile(loss="sparse_categorical_crossentropy",optimizer="SGD",metrics=["accuracy"])
```

```python
history=network.fit(xtrain,ytrain,epochs=25,batch_size=32,validation_data=(xtest,ytest))
```

# 5-Overfitting is clearly present in our suggested model

```python
plt.plot(history.history["loss"],label="train_loss")
plt.plot(history.history["val_loss"],label="test_loss")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

# 6-Let's do data augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
aug=ImageDataGenerator(rotation_range=30,
                      width_shift_range=0.1,
                      height_shift_range=0.1,
                      zoom_range=0.2,
                      horizontal_flip=True,
                      fill_mode="reflect")
history=network.fit_generator(aug.flow(xtrain,ytrain,batch_size=32)
                              ,epochs=20,
                              validation_data=(xtest,ytest),
                              steps_per_epoch=len(xtrain)//32,
                             validation_steps=len(xtest)//32)
```

```python
plt.plot(history.history["loss"],label="train_loss")
plt.plot(history.history["val_loss"],label="test_loss")
plt.legend()
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

# 7-Use pretrained model

### 1-VGG-**16**

```python
from tensorflow.keras.applications import VGG16
```

```python
basemodel=VGG16(weights="imagenet",include_top=False,input_tensor=layers.Input(shape=(32,32,3)))
```

```python
network=models.Sequential([
    basemodel,
    layers.Flatten(),
    layers.Dense(100,activation="relu"),
    layers.Dropout(0.4),
    layers.Dense(17,activation="softmax")
])
```

```python
for layer in basemodel.layers:
    layers.trainable=False
```

```python
network.compile(loss="sparse_categorical_crossentropy",optimizer="SGD",metrics=["accuracy"])
```

```python
history=network.fit(xtrain,ytrain,epochs=25,batch_size=32,validation_data=(xtest,ytest))
```
